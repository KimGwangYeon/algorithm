````python
book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the',
              'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']
# 1번조건
dic ={}
for i in book_title :
    if i in dic :
        dic[i] += 1
    else :
        dic[i] = 1
print(dic)
````

````python
# 2번조건
book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the',
              'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']

book_title_set = set(book_title)
dic = {}
for i in book_title_set :
    dic[i] = book_title.count(i)
print(dic)
````

````python
# 3번조건
# dictionary 안에 값이 있냐 없냐를 보는데 오류안나게 하는 좋은 함수가 get이다.
# if와 자주 쓰이는데 안에 있으면 하고 없으면 else 밑에 명령을 수행한다 라고 하면 된다.

book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the',
              'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']

dic={}
for i in book_title :
    if dic.get(i) :
        dic[i] += 1
    else :
        dic[i] = 1
print(dic)
````

##### 가로세로 높이가 같은 상자들이 쌓여있는 방이 있다. 방이 오른쪽으로 90도 회전하여 상자들이 중력의 영향을 받아 낙하한다고 할 때, 낙차가 가장 큰 상자를 구하여 낙차를 리턴하자.
- 중력은 회전이 완료된 후에 적용된다.
- 상자들은 모두 한쪽 변면에 붙여진 상태로 쌓여 2차원의 형태를 이루며 벽에서 떨어져서 쌓인 상자는 없다.(일직선으로 떨어진다.)
- 방의 가로길이와 세로길이는 9이다.
- 상자의 길이는 가로세로 1이다
- 입력값이 상자가 쌓여진 높이이다.
- 입력값 : [7, 4, 2, 0, 0, 6, 0, 7, 0]
- 출력값 : 7

````python
input = [7, 4, 2, 0, 0, 6, 0, 7, 0]
````

````python
loc = []
for i, value in enumerate(input) :
    for j in range(value) :
        loc.append([i+1, j+1])
print(loc)

rot_loc=[]
for i in loc :
    re_ = [i[1], 10-i[0]]
    rot_loc.append(re_)
print(rot_loc)

locx_locy = {}
for i in rot_loc :
    if i[0]in locx_locy :
        locx_locy[i[0]].append(i[1])
    else :
        locx_locy[i[0]] = [i[1]]
print(locx_locy)
max_height =0    
for i in locx_locy :
    if max(locx_locy[i])-min(locx_locy[i]) > max_height :
        max_height = max(locx_locy[i])-min(locx_locy[i])
        
print(max_height)
````

````python
def is_palin(data) :
    list_s=[]
    for s in data :
        if s.isalnum() :
            list_s.append(s.lower())
    while len(list_s) > 1 :
        if list_s.pop(0) != list_s.pop() :
            print('회문이 아니다!!')
            return
    print('회문이다.')
````

````python
import collections
def is_pal(data) :
    deq = collections.deque()
    for i in data :
        if i.isalnum() :
            deq.append(i.lower())
    while len(deq) > 1 :
        if deq.popleft() != deq.pop() :
            print('회문이 아니다')
            return
    print('회문입니다.')
````
